## Question 2

Explain the relationship between **Primary Keys** and **Foreign Keys**.

Type your answer below this line:
_______________________________________________________________________________________________________________________________

Primary Keys and Foreign Keys are utilized to establish a relationship between two tables in a relational database. The foreign key is a column or set of columns used to reference the primary key column in another table. The primary key is a column or set of columns where each value is unique and identifies a single row of the table.


## Question 3

Explain the different types of relationships entities can have in a SQL database. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________
One-to-one: a single record in one table is related to a single record in another table. One-to-many: a single record in one table is related to one or more records in the second table, but a single record on the second table can only be related to one record in the first table. (This is the most common relationship) Many-to-many: a single record is related to one or more records in the second table and vice versa.

## Question 4

Explain the various types of JOINs possible with SQL. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________
The 4 basic joins are below. Inner Join: returns records with matching values in both tables Left Join: returns all records from the left table and matching records in the right table Right Join: returns all records from the right table and matching records in the left table Outer Join: returns all records in both tables (with matching rows from both sides where available). When there is no match, the missing side will have NULL values.


## Question 5

Explain the relationship between Aggregate functions and GROUP BY statements.

Type your answer below this line:
_______________________________________________________________________________________________________________________________
GROUP BY enables you to group data together based on unique values. Rather than returning every row in a table, unique combinations are returned when values are grouped. It is typically used in conjunction with aggregate functions to summarize values (i.e. SUM, COUNT, AVG, MIN, MAX, etc.)
